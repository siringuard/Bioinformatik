## Aufgabe 2.3 (5/10 Punkte; 1/4 + 0/2 + 4/4)

(a) Beweisen Sie folgende Aussage mittels vollständiger Induktion: Der Algorithmus liefert mit $M = \{1, 5, 10\}$ immer eine optimale Lösung.

**Hinweis:** Betrachten Sie die 3 Fälle $B < 5$, $5 \leq B < 10$ und $10 \leq B$.

a) 

Induktionsbasis:   
$ B < 5 $: $ 1 \in M $ und $ 1 < 5 $
    
$5 \leq B < 10$: $ 5 \in M$ $ 5=5 <10  $

$ 10 \leq B$: $10 \in M$ $ 10 \leq 10$
<br>
<br>
Du musst hier noch begründen, wie M genau aussieht und warum das optimal ist
<br>

Induktionsannahme:

$B+1 < 5$, $5 \leq B+1 < 10$ und $10 \leq B+1$

<br>
Die einzelnen Terme schließen sich gegenseitig aus
<br>

Induktionsschritt:

$ B+1 < 5 $:  $ 1 \in M $ und $ 1+1= 2 < 5 $
    
$5 \leq B+1 < 10$: $ 5 \in M$ $ 5 < 5+1 < 10  $

$ 10 \leq B+1 $: $10 \in M$ $ 10 < 11 = 10+1$

Damit ist die Aussage bewiesen.

Du musst auf die (optimal gelösten) Fälle $B < 5$ und $5\leq B < 10$ verweisen und dadurch die Optimalität begründen

b)

Die $1$-Münze ermöglicht es, jeden Betrag passend zu machen, da sie jede Lücke füllen kann. Indem man zuerst die größte Münze nimmt, reduziert man den verbleibenden Betrag schnell. Solange man immer die größte Münze nehmen kann und am Ende mit $1$-Münzen auffüllt, erreicht man den Betrag mit der minimalen Münzanzahl.
Dies gewährleistet, dass man immer die wenigst möglichen Münzen benutzt, um den Gesamtbetrag zu erreichen und eine optimale Lösung erreicht.

Man kann jeden Betrag passend machen, aber man bekommt nicht unbedingt eine optimale Lösung. Für $M=\{1,3,4\}, B = 6$ wird keine optimale Lösung gefunden. 

c)



```python
def greedy_coin_change(M, B):
    M.sort(reverse=True) 
    result = []
    remaining_amount = B  
    
    for muenze in M:
        while remaining_amount >= muenze:
            result.append(muenze)  
            remaining_amount -= muenze  

    return result 

def test(actual, expected):
    if actual == expected:
        print(f"Test bestanden. Erhalten: {actual}")
    else:
        print(f"Test nicht bestanden. Erwartet: {expected}, erhalten: {actual}")

test(greedy_coin_change([10, 5, 1], 0), [])
test(greedy_coin_change([1, 5, 10], 8), [5, 1, 1, 1])
test(greedy_coin_change([5, 1, 10], 27), [10, 10, 5, 1, 1])
```
